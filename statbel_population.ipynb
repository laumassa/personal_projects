# Belgium Population Trends between 1830 and 2023
This notebook displays static line graphs of Belgian demography, considering births from 1830 to 2023, and deaths from 1840 to 2023.

Datasets are available on Statbel website (Belgian Institute of Statistics), see the following link https://statbel.fgov.be/fr.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_naissances = pd.read_excel("/Users/laure-annemassart/Documents/data_perso/statbel_naissances_MF.xlsx")
df_deces = pd.read_excel("/Users/laure-annemassart/Documents/data_perso/belstat_deces.xlsx")
```

### Historical trends of births

```python
plt.figure(figsize=(20,6))
plt.plot(df_naissances['Year'], df_naissances['Total'], color='limegreen')
plt.xticks(np.arange(1830, 2030, 10))
plt.xlabel('Year')
plt.ylabel('Number of births')
plt.title('Trends in Total Births in Belgium (1830-2023) - All births');
```

```python
plt.figure(figsize=(20,6))
plt.plot(df_naissances['Year'], df_naissances['Garons'], color='deepskyblue')
plt.plot(df_naissances['Year'], df_naissances['Filles'], color='orangered')
plt.legend(['Males', 'Females'])
plt.xticks(np.arange(1830, 2030, 10))
plt.xlabel('Year')
plt.ylabel('Number of births')
plt.title('Trends in Total Births in Belgium (1830-2023)  Male vs. Female births');
```

**Observations:**
- Births peak around 1900: This surge is likely due to a combination of factors, including industrialization, improvements in healthcare, and better sanitary conditions.
- Dramatic decline around World War I, followed by a recovery and increase after the war.
- Second decline around World War II.
- In general, bith rate keeps decreasing since the 60's.

### Historical trends of deaths
NB. The dataset about death rates provides informations starting 1840.

```python
plt.figure(figsize=(20,6))
plt.plot(df_deces['Year'], df_deces['Total'], color='grey')
plt.xticks(np.arange(1840, 2030, 10))
plt.xlabel('Year')
plt.ylabel('Number of deaths')
plt.title('Trends in Total Deaths in Belgium (1840-2023)');
```

**Observations:**
- There are two peaks in mortality around 1845-1850. These can be explained by a combination of factors, including a cholera epidemic, famines in the 1840s, and the early stages of industrialization, which led to poor working conditions and, consequently, a shorter life expectancy.
- Two other mortality peaks are observed around 1870. These coincide with a new cholera epidemic, combined with the rapid progress of urbanization and poor sanitary conditions in major cities.
- A peak during World War I.
- Two peaks during World War II.
- A peak in 2020 during the COVID-19 pandemic.
- In general, mortality rate tends to become more stable since the 1970s, with less dramatic peaks over time, except in 2020.

### Historical trends of birth and deaths (comparative)

```python
plt.figure(figsize=(20,6))
plt.plot(df_naissances['Year'],df_naissances['Total'], color='limegreen')
plt.plot(df_deces['Year'], df_deces['Total'], color='grey')
plt.xticks(np.arange(1830, 2030, 10))
plt.legend(['Births', 'Deaths'])
plt.xlabel('Year')
plt.ylabel('Number')
plt.title('Total Number of Births and Deaths in Belgium (1830-2023)');
```

**Additional observations:**

There are 3 periods during which the death curve significatively exceeds the birth curve: World War I, World War II and the COVID-19 pandemics.
